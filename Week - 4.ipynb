<h3> Profile in R</h3>

```r
# understand why our program is taking a long time.
# collect data, and profile, don't guess!!!

# 1. system.time()
# two times of time we'll get.
# a. user time (CPU spent)
# b. elapsed time (we experience)

system.time(readLines("http://www.jhsph.edu"))
```

```r
# use multi core, split computation across two cores.
str(svd)
hilbert = function(n){
    i = 1:n
    1 / outer(i-1, i, "+")
}

x = hilbert(1000)
system.time(svd(x))
```

```r
system.time({
    n = 1000
    r = numeric(n)
    for(i in 1:n){
        x = rnorm(n)
        r[i] = mean(x)
    }
})
```
